```python
import pandas as pd
df=pd.read_csv('wine.csv')
df
```

```python
df.isna().sum()
#check null values 
```

```python
for col in df.columns:
    if df[col].isna().sum()>0:
        df[col]=df[col].fillna(df[col].mean())
df.isna().sum()
#if there is any null values we replace them with mean value.
```

```python
df.describe()
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
pd.plotting.scatter_matrix(df,figsize=(20,20))
plt.show()
#it is the diagram of -->df
```

```python
pd.get_dummies(df,columns=['type'],drop_first=True)
#we will convert  'type' column into numeric as you can see
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

x=df.drop('quality',axis=1)
y=df['quality']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
#now we predict the value.

y_pred=model.predict(x_test)
```

```python
print("co-efficient:",model.coef_)
#coefficient
print("intercept:",model.intercept_).
#intercept of model
```

```python
print("mean sqaured error :",mean_squared_error(y_test,y_pred))
```

```python
print("mean absolute error :",mean_absolute_error(y_test,y_pred))
```

```python
print("r2 score :",r2_score(y_test,y_pred))
```
